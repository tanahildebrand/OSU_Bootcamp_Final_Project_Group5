```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(Path('../happiness_kcal_by_country.csv'))
df.head()
```

```python
df.dtypes
```

```python
plt.scatter(df.Alcoholic_Beverages, df.happiness_score)

font1 = {'family':'calibri','color':'blue','size':15}
font2 = {'family':'calibri','color':'darkred','size':12}

plt.title('Alcoholic Beverages vs Happiness Score', fontdict = font1)
plt.xlabel('Alcoholic Beverages',fontdict = font2)
plt.ylabel('Happiness Score',fontdict = font2)

plt.show()
```
